Dataset Description


The dataset provides a comprehensive repository of Formula 1 (F1) World Championship üèÜdata essential for hosting a machine learning competition. It includes detailed information on race results, driver statistics, constructor details, and race statuses spanning multiple decades of F1 history.

This dataset is invaluable for researchers and participants to analyze performance trends, predict race outcomes, model driver and constructor standings, and explore the technical nuances of F1 racing. By leveraging this extensive dataset, participants can develop machine learning models that uncover evolving strategies, adapt to changing technologies, and predict the impact of regulatory changes on race dynamics.

This resource enables a deep dive into the complexities and developments within the world of Formula 1, offering a robust foundation for competitive analysis and innovation in machine learning applications within motorsport.

```python
##Influential points(outlier detection) using z score method
from scipy import stats
import numpy as np
# Calculate Z-scores for numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
z_scores = stats.zscore(df[numerical_columns])

# Find indices where Z-score exceeds threshold (e.g., abs(Z-score) > 3 for significant outliers)
threshold = 3
outlier_indices = (np.abs(z_scores) > threshold).any(axis=1)

# Print influential points
influential_points = df[outlier_indices]
print("Influential points detected using Z-score:")
print(influential_points)
```

```python
# Calculate IQR for numerical columns
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Determine outliers based on IQR
outlier_indices = ((df[numerical_columns] < (Q1 - 1.5 * IQR)) | (df[numerical_columns] > (Q3 + 1.5 * IQR))).any(axis=1)

# Print influential points
influential_points = df[outlier_indices]
print("Influential points detected using IQR:")
print(influential_points)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Assuming 'position' is your target variable and other columns are predictors
X = df.drop(columns=['position'])
y = df['position'].astype(float)  # Ensure 'position' is in float or numeric format

# Fit a simple linear regression model
X = sm.add_constant(X)  # Adding constant for intercept
model = sm.OLS(y, X).fit()

# Calculate leverage and residuals
leverage = model.get_influence().hat_matrix_diag
residuals = model.get_influence().resid_studentized_external

# Identify influential points based on leverage and residuals
influential_points = df[(np.abs(residuals) > 2) | (leverage > 2 * (X.shape[1] / X.shape[0]))]
print("Influential points detected using model residuals and leverage:")
print(influential_points)
```
